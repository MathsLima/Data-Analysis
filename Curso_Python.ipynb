Estudo Hands On de uma análise e criação de indicadores de vendas de produtos eletrônicos utilizando arquivos csv e linguagem de programação python.

### **Importando as listas de dados**

```python
import os
import pandas as pd

lista_arquivos = os.listdir("/content/drive/MyDrive/Curso_Python/Vendas")
display(lista_arquivos)
```

### **Importando e criando a base de dados**

```python
tabela_total = pd.DataFrame()

# Importando as bases de dados de vendas
for arquivo in lista_arquivos:
  if "Vendas" in arquivo:
  
# Importando o arquivo
    tabela = pd.read_csv(f"/content/drive/MyDrive/Curso_Python/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela)

# Compilaçao das bases de dados
display(tabela_total)    
```

###**Indicadores**


```python
# Produto mais vendido por quantidade

tabela_produtos = tabela_total.groupby("Produto").sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending = False)
display(tabela_produtos)
```

```python
# Calculo de produto que mais faturou

tabela_total ["Faturamento"] = tabela_total["Quantidade Vendida"] * tabela_total["Preco Unitario"]

tabela_faturamento = tabela_total.groupby("Produto").sum()
tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by="Faturamento", ascending = False)
display(tabela_faturamento)
```

```python
# Loja que mais vendeu em faturamento

tabela_lojas = tabela_total.groupby("Loja").sum()
tabela_lojas = tabela_lojas[["Faturamento"]].sort_values(by="Faturamento", ascending = False)
display(tabela_lojas)
```

### **Gráficos**

```python
import plotly.express as px

# Produto mais vendido
grafico_qtd = px.bar(tabela_produtos, x=tabela_produtos.index, y="Quantidade Vendida")
grafico_qtd.show()

# Produto mais faturado
grafico_prod = px.bar(tabela_faturamento, x=tabela_faturamento.index, y="Faturamento")
grafico_prod.show()

# Loja com maior faturamanto
grafico_fat = px.bar(tabela_lojas, x=tabela_lojas.index, y="Faturamento")
grafico_fat.show()
```
