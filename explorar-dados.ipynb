EXPLORAR DADOS DE UM DATAFRAME  

```python
import pandas as pd
import numpy as np

dados = [50,50,47,97,49,3,53,42,26,74,82,62,37,15,70,27,36,35,48,52,63,64]
notas = np.array(dados)

horas_studo= [10.0,11.5,9.0,16.0,9.25,1.0,11.5,9.0,8.5,14.5,15.5,
               13.75,9.0,8.0,15.5,8.0,9.0,6.0,10.0,12.0,12.5,12.0]

dados_estudante = np.array([horas_studo, notas])

df_estudantes = pd.DataFrame({'Nome': ['Dan', 'Joann', 'Pedro', 'Rosie', 'Ethan', 'Vicky', 'Frederic', 'Jimmie', 
                                     'Rhonda', 'Giovanni', 'Francesca', 'Rajab', 'Naiyana', 'Kian', 'Jenny',
                                     'Jakeem','Helena','Ismat','Anila','Skye','Daniel','Aisha'],
                            'Horasestudos':dados_estudante[0],
                            'Notas':dados_estudante[1]})

df_estudantes
```

```python
# media das horas de estudo usando o nome da coluna como indice
media_estudos = df_estudantes['Horasestudos'].mean()

# media das notas usando o nome da coluna como indice
media_notas = df_estudantes['Notas'].mean()

print('A media de horas de estudos é: {:.2f} \nA media de notas é {:.2f}'.format(media_estudos, media_notas))
```

```python
# localizar valores acima da media
df_estudantes[df_estudantes.Horasestudos > media_estudos]
```

```python
# media sobre a media 
df_estudantes[df_estudantes.Horasestudos > media_estudos].Notas.mean()
```

```python
# adicionando uma nova coluna comparando com os valores da tabela
passar = pd.Series(df_estudantes['Notas'] >= 60)
df_estudantes = pd.concat([df_estudantes, passar.rename('Passou')], axis=1)
df_estudantes
```

```python
# agrupando dados 
print(df_estudantes.groupby(df_estudantes.Passou).Nome.count())
```

```python
# media de estudos e notas para aprovados e reprovados
print(df_estudantes.groupby(df_estudantes.Passou)['Horasestudos', 'Notas'].mean())
```

```python
# colocar em ordem crescente
df_estudantes  = df_estudantes.sort_values('Notas', ascending=False)
df_estudantes
```
