paginando o repositorio

```python
repos_list = [] #lista para armazenar as paginas 
for page_num in range(1, 6): #laço para percorrer as paginas
    try:
        url_page = f'{url}?page={page_num}'  
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

transformando os dados

```python
repos_list[0][2] #zero e a pagina e 2 o repositorio
```

```python
repos_list[0][2]['name'] #selecionando o nome do repositorio
```

```python
# obtendo todos os nomes dos repositorios
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

obtendo os dados de qual linguagem é utilizada no repositorio

```python
repos_list[1][1]['language']
```

```python
repos_language = []
for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
```

criando o dataframe

```python
amazon_data = pd.DataFrame()
amazon_data['repository_name'] = repos_name
amazon_data['language'] = repos_language
```
