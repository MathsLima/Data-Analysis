```python
import pandas as pd
```

configurando as opções do dataframe

```python
pd.get_option("display.max_rows")
pd.get_option("display.max_columns")
```

carregando os dados

```python
bairros = pd.read_excel(
    io = "bairros.xlsx",
    sheet_name="Preço médio por tipo",
    usecols="C:E", #selecionando intervalo de colunas a ser utilizadas
    header = 2, #selecionando a linha que inicia a tabela
    index_col=[0, 1],
    names=['bairros', 'tipo', 'valor_m2_bairro']
)
bairros
```

```python
#carregando o json
dados = pd.read_json(
    path_or_buf='realestates.json',
    orient='columns'
)
dados
```

tratando o json

```python
#selecionando os dados da tabela normal do json - método 1
dados_normal = pd.json_normalize(data = dados.normal)
dados_normal_listings = pd.json_normalize(data = dados_normal['listings'].iloc[0], sep='_' )  #sep serve para alterar o header
dados_normal_listings.head()
```

```python
#selecionando os dados da tabela normal do json - método 2
dados_normal_listings = pd.json_normalize(data=dados.normal, sep='_', record_path = ['listings'])
dados_normal_listings.head()
```

```python
#selecionando os dados da tabela highlights do json - método 2
dados_highlights_listings = pd.json_normalize(data=dados.highlights, sep='_', record_path = ['listings'])
dados_highlights_listings.head()
```

 ### trabalhando com dado textual

Transformando string em lista no dataframe

```python
dados_normal_listings.head(2)
```

```python
#selcionando os itens de dentro da lista - método 1
lista_str = dados_normal_listings.loc[0, 'imovel_caracteristicas_propriedade']
lista_str[1:-1] 
```

```python
#selcionando os itens de dentro da lista - método 2
lista_str = dados_normal_listings.loc[0, 'imovel_caracteristicas_propriedade']
lista_str.strip("[]")
```

```python
#eliminando as aspas
lista = lista_str[1:-1].replace("'"," ").split(", ") 
lista
#replace é para trocar as aspas duplas por nada 
#cada vez que encontrar a virgula e espaço é colocado o argumento em uma lista
```

```python
texto = dados_normal_listings.loc[:, 'imovel_caracteristicas_propriedade']
texto.str[1:-1] #remove []
texto.str[1:-1].replace("'", "") #remover as aspas
texto = texto.str[1:-1].str.replace("'", "") #separa por virgulas
texto
# .loc pega os itens da coluna
# .loc[:] pega todos os itens da coluna
# .loc[:, 'header'] pega todos os itens de uma coluna especifica
```

Filter

```python
columns = dados_normal_listings.filter(like='imovel_caracteristicas').columns

for column in columns:
    dados_normal_listings[column] = dados_normal_listings[column].str[1:-1].str.replace("'", "").str.split(", ")
    dados_highlights_listings[column] = dados_highlights_listings[column].str[1:-1].str.replace("'", "").str.split(", ")
```

```python
dados_normal_listings.loc[0, 'imovel_caracteristicas_propriedade'][0]
```

### criando as colunas quartos, suites e banheiros

```python
dados_listings['anuncio_descricao'].values[:10]
```

str.extractall

```python
configuracao = dados_listings['anuncio_descricao'].str.extractall('(\d+)')
configuracao.head(9)
```

unstack

droplevel

```python
#inserindo os valores no dataframe geral
dados_listings = pd.merge(dados_listings, configuracao, left_index=True, right_index=True)
dados_listings
```

com dados numericos - cut

```python
valor_minimo = dados_listings['anuncio_valores_venda'].min()
```

```python
valor_maximo = dados_listings['anuncio_valores_venda'].max()
```

```python
rotulos = ['Popular', 'Padrão', 'Alto Padrão']
```

utilizando classe fixa

```python
pd.cut(x=dados_listings['anuncio_valores_venda'], bins=3, labels=rotulos)
```

utilizando classe personalizada

```python
# É preciso N pontos para obter N - 1 classes
classes = [valor_minimo, 400000, 2000000, valor_maximo]
```

```python
pd.cut(x=dados_listings['anuncio_valores_venda'], bins=classes, labels=rotulos, include_lowest=True)
```

```python
dados_listings['classe_valor'] = pd.cut(x=dados_listings['anuncio_valores_venda'], bins=classes, labels=rotulos, include_lowest=True)
dados_listings
```

pivot

```python
dados_listings.pivot_table(
    values='valor_m2',
     index='imovel_tipos_propriedade',
    columns='classe_valor',
    fill_value='-', #para preencher os valores nulos
    margins=True,
    margins_name='Média Geral',
    aggfunc='mean'
)
```

```python
dados_listings.pivot_table(
   values='valor_m2',
     index=['imovel_tipos_propriedade', 'classe_valor'],
     columns=['anuncio_tipos_listagem','Piscina'],
     fill_value='-',
     margins=True,
     margins_name='Média Geral',
     dropna=False,
     aggfunc='mean'

)
```

##### tabulaçao partindo de infromaçao no formato de lista

```python
caracteristicas_propriedade = dados_listings['imovel_caracteristicas_propriedade']
caracteristicas_propriedade
```

```python
caracteristicas_propriedade_itens = caracteristicas_propriedade.explode()
caracteristicas_propriedade_itens
```
